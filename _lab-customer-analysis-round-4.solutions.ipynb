```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
number_of_response    = len(mktdf[mktdf['Response']=='Yes'])
number_of_no_response = len(mktdf) - number_of_response

x = [number_of_response, number_of_no_response]
labels = ['responded', 'not responded']
colors = ['pink', 'violet']

fig, ax = plt.subplots()
ax.pie(x, colors=colors, labels=labels, autopct='%.1f%%')
ax.set_title('Reponses')

plt.show()
```
